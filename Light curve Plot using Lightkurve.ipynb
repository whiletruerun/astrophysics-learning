# PLOTTING LIGHTKURVES

## What I'm Learning

Light curves are plots of a star's brightness (flux) versus time.

Sometimes, periodic dips in brightness can occur when a planet passes in front of the star from our point of view. This is known as a transit. By analyzing these dips, we can estimate properties such as the planet's size (from the depth of the dip) and its orbital period (from how often the dips happen).

So here I tried to visualize and interpret light curve data using python and the Lightkurve library

```python
try:
    import lightkurve
except ImportError:
    !pip install lightkurve
```

```python
TIC = 'TIC 100100827'
```

```python
sector_data = lightkurve.search_lightcurve(TIC, author='SPOC')
print(sector_data)
```

```python
sector_data = lightkurve.search_lightcurve(TIC, author='SPOC',sector=69,exptime=20)
print(sector_data)
```

```python
lc = sector_data.download()
```

```python
lc.plot(marker='.', linewidth = 0, alpha = 0.3)
```
